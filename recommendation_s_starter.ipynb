```python
# **WE ARE MAKING CONTENT BASED RECOMMENDATION SYSTEM**

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
##Step 1: Read CSV File
df=pd.read_csv("https://raw.githubusercontent.com/codeheroku/Introduction-to-Machine-Learning/master/Building%20a%20Movie%20Recommendation%20Engine/movie_dataset.csv")
df.head()
##Step 2: Select Features

##Step 3: Create a column in DF which combines all selected features

##Step 4: Create count matrix from this new combined column

##Step 5: Compute the Cosine Similarity based on the count_matrix

movie_user_likes = "Avatar"

## Step 6: Get index of this movie from its title

## Step 7: Get a list of similar movies in descending order of similarity score


## Step 8: Print titles of first 50 movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv=CountVectorizer()   #Its a class so we have to create its object
```

```python
def combine_features(row):
    
    return (row["keywords"] +" "+row["cast"]+" "+row["genres"] + ""+ row["director"])
```

* Now, we need to call this function over each row of our dataframe. But, before doing that, we need to clean and preprocess the data for our use. We will fill all the NaN values with blank string in the dataframe.

```python
features=["keywords","cast","genres","director"] #put all categorical data it this list
for i in features:
    df[i]=df[i].fillna("") #Fill all missing values as "missing".
```

```python
df["combined_features"]=df.apply(combine_features,axis=1)
```

* Now that we have obtained the combined strings, we can now feed these strings to a CountVectorizer() object for getting the count matrix.

```python
cv_fit=cv.fit_transform(df["combined_features"]) #feeding combined strings(movie contents) to CountVectorizer() object
```

```python
cos_sim=cosine_similarity(cv_fit)
```

```python
###### helper functions. Use them when needed #######
def get_title_from_index(index):
	return df[df.index == index]["title"][0]

def get_index_from_title(title):
	return df[df.title == title]["index"].values[0]
##################################################
```

* Our next step is to get the title of the movie that the user currently likes. Then we will find the index of that movie. After that, we will access the row corresponding to this movie in the similarity matrix. Thus, we will get the similarity scores of all other movies from the current movie. Then we will enumerate through all the similarity scores of that movie to make a tuple of movie index and similarity score. This will convert a row of similarity scores like this- [1 0.5 0.2 0.9] to this- [(0, 1) (1, 0.5) (2, 0.2) (3, 0.9)] . Here, each item is in this form- (movie index, similarity score).

```python
movie_user_likes
```

```python
movie_index=get_index_from_title(movie_user_likes)
similar_movies=list(enumerate(cos_sim[movie_index]))#accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
```

```python
movie_index
```

```python
similar_movies # it shows how "avatar" is similar to movies (movies at that index,cosine_similarity)
```

```python
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
sorted_similar_movies
```

```python
i=0
print("Top 5 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```
